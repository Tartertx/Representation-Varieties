```python
import pandas as pd
import random as rd

t = 4
p = 5
def getWeights(t,p):
    weights = []
    inverses = []
    for i in range(0,t):
        # weights +=[1]
        weights += [rd.randint(1,p-1)]
        inverses += [pow(weights[i],-1,p)]
    return weights, inverses

def gammaExp(weights, inverses):
    count = 1
    for i in range(0,t-1):
        count += inverses[i] * weights[0]
    return count

weights, inverses = getWeights(t,p)
count = gammaExp(weights, inverses)
gammaprod = [f"g_1^({count})"]

def lij(i,j,t,weights):
    lijval = -1 * weights[0] * (inverses[j-1] - i)
    for i in range(j, t):
        lijval += -1 * weights[0] * inverses[i]
    return lijval

print(f'Weights: {weights}')
print(f'Inverses: {inverses}')

for j in range(1,t):
    for i in range(0,inverses[j]):
        gammaprod += [f"a^{-1}_({j}, {lij(i,j,t,weights)})"]        


print(gammaprod)
```
